# One Dimensional plots

 one continious variable 

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(mpg['hwy'], color="orange", bins=6)
plt.title("Highway Mileage")
plt.xlabel("Mileage")
plt.ylabel("Frequency")
plt.show()
```

```python
#boxplot
plt.boxplot(mpg['hwy'])
plt.title("Boxplot of Highway Mileage")
plt.xlabel("Cars")
plt.ylabel("Mileage")
plt.show()
```

## Bar chart for discrete variable

```python
#barchart
data = mpg['class'].value_counts()
plt.bar(data.index,data.values, width = 0.5)
plt.title("Barchart of Vechicles count by Class")
plt.xlabel("Class")
plt.ylabel("Frequency")
plt.show()
```

## Two dimensional 

```python
plt.scatter(mpg["cty"], mpg["hwy"] , color = "red", marker="+")
plt.title("City Vs Highway comparison")
plt.xlabel(" City Mileage")
plt.ylabel("Highway Mileage")
plt.show()
```

## one discrete and one cont 

```python
plt.boxplot(x=mpg["hwy"])
plt.show()
```

```python
## one discrete and one continious variable -- Boxplot (pandas)
## using pandas plotting 
boxplot = mpg.boxplot(column="hwy", by = "class")
```

## two discrete variables 


```python
## using Pandas plotting 
data = pd.crosstab(mpg["class"], mpg["drv"])
df = pd.DataFrame(data)
df.plot.bar()
```

```python
data = pd.crosstab(mpg["class"], mpg["drv"])
data
```

```python
## using Matplotlib 
data = pd.crosstab(mpg["class"], mpg["drv"])
#df = pd.DataFrame(data, index = data.index)
plt.bar(data.index, data['4'])
plt.bar(data.index, data['f'],bottom=data['4'], color = "green" )
plt.bar(data.index, data['r'], bottom = data['4'], color = "red")
plt.title(" Count of vehicles by Class")
plt.xlabel("Class of Vehicle")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
# create figure and axis
fig, ax = plt.subplots()
# plot histogram
ax.hist(mpg['cty'])
# set title and labels
ax.set_title('City Mileage')
ax.set_xlabel('Mileage')
ax.set_ylabel('Frequency')
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10,5))
ax[0].hist(mpg["hwy"])
ax[1].boxplot(mpg["hwy"])
ax[0].set_title("Hitogram of Higway Mileage")
ax[1].set_title("Boxplot of Highway Mileage")
ax[0].set_xlabel("High Mileage")
ax[0].set_ylabel("Frequency")
ax[1].set_xlabel("boxplot")
ax[1].set_ylabel("Mileage")
```

```python
## barchart for one discrete avriable -- auto color selection 
sns.countplot(mpg["class"])
```

```python
## Scatter plot  for two continouos variable 
sns.regplot(x = "cty", y = "hwy", data = mpg, fit_reg=False) ## fit_reg = False will ignore the line 
```

```python
## boxplots for one cont and one discrete variable 

sns.boxplot(x = "class", y = "hwy", data = mpg )
```

```python
## barchart for two discrete variables -- default dodged bar chart 

sns.countplot(x = "class", hue = "drv", data = mpg)
```

```python
## Stacked barchart 
sns.countplot(x = "class", hue = "drv", data = mpg, dodge=False, )
```
